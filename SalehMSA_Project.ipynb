```python
islamic= float(input("enter islamic mark :"))
arabic= float(input("enter arabic mark :"))
english= float(input("enter english mark :"))
marks = [0,0,0,0,0,0,]
marks[0]= islamic
marks[1]= arabic
marks[2]= english
print("**************************************")

print(''' choose at track from the next menu : 
1- enter a to select advanced
2- enter b to select foundation ''')

track = input("enter your track :")
if track == 'a':
    physics = float(input("enter physics mark :"))
    chemistry = float(input("enter chemistry mark :"))
    marks[3] = physics
    marks[4] = chemistry
    
    print(''' choose at track from the next menu : 
    1- enter a to select engineering
    2- enter b to select medical ''')
    
    track = input("enter your track :")
    if track == "e":
        math = float(input("enter math mark :"))
        marks[5] = math
        total= sum(marks)
    else:
        biology = float(input("enter biology mark :"))
        marks[5] = biology
        total= sum(marks)    
    
else :
    geographic = float(input("enter geographic mark :"))
    history = float(input("enter history mark :"))
    comupter = float(input("enter comupter mark :"))
    marks[3]= geographic
    marks[4]= history
    marks[5]= comupter
    total= sum(marks)
print("____________________________________________")    

average = total/6
print("The totol is :", total )
print("The average is :", average )


if average >= 50 :
    print("the result is (PASS)")
    if average >=90:
        print("the grade is ( A )")
    elif average >= 80:
        print("the grade is ( B )")
    elif average >= 70:
        print("the grade is ( C )")
    elif average >= 60:
        print("the grade is ( D )")
else:
    print("the result is (Fail)")   
```
